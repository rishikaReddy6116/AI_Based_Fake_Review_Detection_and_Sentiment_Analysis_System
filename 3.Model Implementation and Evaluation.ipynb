```python
# FINAL MODEL FOR DEPLOYMENT
import pickle
import re
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from scipy.sparse import hstack

nltk.download('stopwords')
sw = set(stopwords.words('english'))
stemmer = PorterStemmer()

# CLEAN TEXT
def clean_text(text):
    text = re.sub("[^a-zA-Z]", " ", str(text))
    text = text.lower()
    words = text.split()
    words = [stemmer.stem(word) for word in words if word not in sw]
    return " ".join(words)

# LOAD DATA
df = pd.read_csv("data and pickle files/cleaned_data.csv", encoding="latin1")
df = df.dropna(subset=['review_text'])

df['cleaned_review'] = df['review_text'].apply(clean_text)

# EXTRA FEATURES
def extract_features(text):
    text = str(text)
    words = text.split()
    return pd.Series({
        'length': len(text),
        'exclamation_count': text.count('!'),
        'capital_count': sum(1 for c in text if c.isupper()),
        'word_count': len(words),
        'unique_words': len(set(words)),
        'unique_ratio': len(set(words))/len(words) if len(words)>0 else 0,
        'repeated_words': len(words)-len(set(words))
    })

feature_df = df['review_text'].apply(extract_features)
df = pd.concat([df, feature_df], axis=1)

# CREATE LABEL


X = df['cleaned_review']
# CREATE FAKE REVIEW LABEL
df['fake_label'] = 0

df.loc[df['repeated_words'] > 2, 'fake_label'] = 1
df.loc[df['exclamation_count'] > 2, 'fake_label'] = 1
df.loc[df['capital_count'] > 5, 'fake_label'] = 1

print(df['fake_label'].value_counts())

# SPLIT
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# TFIDF
tfidf = TfidfVectorizer(stop_words='english')
X_train_text = tfidf.fit_transform(X_train)
X_test_text = tfidf.transform(X_test)

feature_columns = ['length','exclamation_count','capital_count','word_count','unique_words','unique_ratio','repeated_words']
X_train_feat = df.loc[X_train.index, feature_columns].values

X_train_final = hstack([X_train_text, X_train_feat])

# TRAIN MODEL
model = LogisticRegression()
model.fit(X_train_final, y_train)

# SAVE FILES
pickle.dump(model, open('data and pickle files/best_model.pkl','wb'))
pickle.dump(tfidf, open('data and pickle files/tfidf.pkl','wb'))
pickle.dump(feature_columns, open('data and pickle files/feature_columns.pkl','wb'))

print("FINAL MODEL SAVED FOR DEPLOYMENT")
```
